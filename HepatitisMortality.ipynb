### Model Building

```python

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

# Metrics
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# Features for Initial Data
df.columns
```

```python
xfeatures
```

```python
ylabels = df['class']
```

```python
# Selected Features
df.columns
```

```python
xfeatures_best = df[['age', 'sex', 'steroid', 'antivirals','fatigue','spiders',
       'ascites', 'varices', 'bilirubin', 'alk_phosphate', 'sgot', 'albumin',
       'protime', 'histology']]
```

```python
# train /test dataset
x_train,x_test,y_train,y_test = train_test_split(xfeatures,ylabels,test_size=0.30,random_state=9)
```

```python
# train /test dataset for best features
x_train_b,x_test_b,y_train_b,y_test_b = train_test_split(xfeatures_best,ylabels,test_size=0.30,random_state=7)
```

```python
# Building Model
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
# Model Accuracy
# Method 1
logreg.score(x_test,y_test)
```

```python

logreg.predict(x_test)
```

```python
print(list(y_test))
```

```python
#Accuracy score 85%
accuracy_score(y_test,logreg.predict(x_test))
```

```python

model_logit = LogisticRegression()
model_logit.fit(x_train_b,y_train_b)
```

```python

model_logit.score(x_test_b,y_test_b)
```
